```python
import pandas as pd
df=pd.read_csv('aapl_no_dates.csv')
df.head(8)
```

```python
range=pd.date_range(start='6/1/2017',end='6/30/2017',freq='B')
range
```

```python
df.set_index(range,inplace=True)
df
```

```python
range=pd.date_range(start='6/1/2017',end='6/30/2017',freq='D')
range
```

```python
df.set_index(range,inplace=True)
df
```
