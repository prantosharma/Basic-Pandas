```python
import pandas as pd
df1=pd.DataFrame({'A':[10],
                 'B':[20],
                 'C':[30]
    })
df1
```

```python
df2=pd.DataFrame({'D':[40,12],
                 'E':[50,15],
                 'F':[60,18]
    })
df2

```

```python
n_df=pd.concat([df1,df2])
n_df
```

```python
bangladesh_weather=pd.DataFrame({
    'city':['dh','kh','ctg'],
    'temp':[25,29,33]
})
bangladesh_weather
```

```python
us_weather=pd.DataFrame({
    'city':['dhk','khl','ctga'],
    'temp':[55,39,35]
})
us_weather
```

```python
n_df=pd.concat([bangladesh_weather,us_weather])
n_df
```

```python
n_df=pd.concat([bangladesh_weather,us_weather],ignore_index=True)
n_df
```

```python
n_df=pd.concat([bangladesh_weather,us_weather],keys=['bangladesh_weather','us_weather'])
n_df
```

```python
n_df=pd.concat([bangladesh_weather,us_weather],
               keys=['bangladesh_weather','us_weather'],names=['country names','serial no'])
n_df
```

```python
n_df.loc['us_weather']
```

```python
temp=pd.DataFrame({
    'city':['dhk','khl','ctga'],
    'temperature':[31,35,36]
})
temp
```

```python
wind=pd.DataFrame({
    'city':['dhk','khl','ctga'],
    'windspeed':[5,9,6]
})
wind
```

```python
new_df=pd.concat([temp,wind],axis=1)
new_df
```

```python
temp=pd.DataFrame({
    'city':['dhk','khl','ctga'],
    'temperature':[31,35,36]
},index=[1,2,3])
temp

wind=pd.DataFrame({
    'city':['khl','dhk','ctga'],
    'windspeed':[5,9,6]
},index=[2,1,3])
wind
```

```python
nw_df=pd.concat([temp,wind],axis=1)
nw_df
```

```python
n=pd.Series(['rain','snow','sunny'],name='event')
n
```

```python
nw_df=pd.concat([temp,wind,n],axis=1,skip=0)
nw_df
```
